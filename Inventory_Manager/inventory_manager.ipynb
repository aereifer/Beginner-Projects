Challenge: Inventory Management System
Scenario: you are tasked with building a basic inventory management system using python dictionaries

Requirements:
1. Create and initial inventory dictionary: Start with an empty dictionary to store your inventory
2. Add items to the inventory: Write functions to:
   ~Add a new item to the inventory with its name, quantity, and price per unit
   ~If the item already exists, update its quantity and price
3. Remove items from the inventory: Write a function to:
   ~Remove an item from the inventory by its name
   ~If the item doesn't exist, display an appropriate message
4. Update inventory quantities: Write a function to:
   ~Increase or decrease the quantity of an existing item
5. Generate an inventory report: Write a function to:
   ~Display a formatted table showing all items in the inventory, their quantities, price per unit, and total value (quantity * price)

```python
#METHODS
def task_num_check(num):
    while num not in range(0,7):
        if num.isdigit() == True:
            num = int(num)
        
            if num in range(0,7):
                pass
            else:
                num = input("Invalid. The number you entered was out of range. Please enter a number. (1-6)")
        else:
            num = input("Please enter a number. (1-6)")
            
    return num


def is_positive(num):
    try:
        num > 0
        return True
    except:
        return False

    
def float_check(num):
    try:
        float(num)
        return True
    except:
        return False


def int_check(num):
    while not num.isdigit() or is_positive(num) != True:
        if num.isdigit():
            num = int(num)

            if num > 0:
                break
            else:
                num = input("The number you entered needs to be greater than 0. Please enter another number.")
        else:
            num = input("Invalid input. Please enter a a positive integer.")

    return num



def price_check(num):
    while float_check(num) != True or is_positive(num) != True:
        if float_check(num.lstrip("$")) == True:
            num = float(num.lstrip("$"))

            if float(num) > 0:
                break
            else:
                num = input("Invalid input. Please enter a price that is greater than 0.")
        else:
            num = input("Invalid input. Please enter a number.")

    return ("${:.2f}".format(num))


def view_task_list():
    print("1. View inventory")
    print("2. Add item to inventory")
    print("3. Remove item from inventory")
    print("4. Update inventory quantities")
    print("5. Generate inventory report")
    print("6. Exit")
    print("\n")


def view_dicts(quantity_dict, prices_dict):
    for key, value in quantity_dict.items():
        print(f"Item: {key}")
        print(f"Quantity: {value}")
        print(f"{prices_dict[key]} per unit")
        print("\n")


def add_item(name, quantity, price, quantity_dict, prices_dict):
    if name in quantity_dict.keys():
        quantity_dict[name] = quantity_dict.values() + quantity_dict[name]
    else:
        quantity_dict[name] = quantity
        prices_dict[name] = price


def remove_item(name, quantity_dict, prices_dict):
    if name in quantity_dict.keys():
        del quantity_dict[name]
        del prices_dict[name]
    else:
        print("The item '{}' is not in your inventory.".format(name))


def update_quantity(name, quantity_dict):
    if name in quantity_dict.keys():
        quantity_dict[name] = int_check(input("Enter the new quantity."))
    else:
        print("The item '{}' is not in your inventory.".format(name))


def gen_report(quantity_dict, prices_dict):
    for key, value in quantity_dict.items():  
        print("Item: {}".format(key))
        print("Quantity: {}".format(value))
        print("Price: {}".format(prices_dict[key]))
        print("Total Value: ${:.2f}".format(value * float(prices_dict[key].lstrip("$"))))
        print("\n")        
```

```python
quantity_dict = {}
prices_dict = {}
task_num = 0

print("Welcome to the Inventory Management System.")
print("\n")

view_task_list()
task_num = task_num_check(input("Enter the number of the task you would like to perform. (1-6)"))

while task_num != 6:
    if task_num == 1:
        view_dicts(quantity_dict, prices_dict)
    elif task_num == 2:
        add_item(input("Enter the name of the item you would like to add."), int_check(input("Enter the quanity.")), price_check(input("Enter the price per unit.")), quantity_dict, prices_dict)
    elif task_num == 3:
        remove_item(input("Enter the name of the item you would like to remove."), quantity_dict, prices_dict)
    elif task_num == 4:
        update_quantity(input("Enter the name of the item you would like to change the quantity of."), quantity_dict)
    else:
        gen_report(quantity_dict, prices_dict)

    task_num = task_num_check(input("Enter the number of the next task you would like to perform. (1-6)"))
                            
```
